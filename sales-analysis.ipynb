```python
# Importing the tools we need
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Makes charts look clean and big
plt.rcParams['figure.figsize'] = (12, 5)
sns.set_style("whitegrid")

print("Libraries loaded successfully!")
```

```python
# Load the CSV file into a table called 'df' (dataframe)
df = pd.read_csv('Sample - Superstore.csv', encoding='latin-1')

# Show the first 5 rows so we can see what the data looks like
print("Shape of data:", df.shape)
print("\nFirst 5 rows:")
df.head()
```

```python
# Group all rows by product name, add up their sales, take top 10
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

# Draw a horizontal bar chart
plt.figure()
top_products.plot(kind='barh', color='steelblue')
plt.title('Top 10 Products by Sales')
plt.xlabel('Total Sales ($)')
plt.tight_layout()
plt.savefig('top_products.png')  # saves the chart as an image
plt.show()
print("Chart saved!")
```

```python
# Group by Region, sum the sales
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)

plt.figure()
region_sales.plot(kind='bar', color=['#2196F3','#4CAF50','#FF9800','#E91E63'])
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('region_sales.png')
plt.show()

# Group by Category
category_sales = df.groupby('Category')['Sales'].sum()

plt.figure()
plt.pie(category_sales, labels=category_sales.index, autopct='%1.1f%%', startangle=90)
plt.title('Sales Share by Category')
plt.tight_layout()
plt.savefig('category_sales.png')
plt.show()

print("All charts done! Your project is complete.")
```
